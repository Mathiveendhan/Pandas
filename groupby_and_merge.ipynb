```python
df1=pd.DataFrame({'roll.no':[1,2,3,4,5],'maths':[77,65,78,89,90]})
df1
```

```python
df3=pd.DataFrame({'roll.no':[1,2,3,8,9],'computer':[77,65,78,89,90]})
df4=pd.DataFrame({'roll.no':[1,2,3,6,7],'science':[77,65,78,89,90]})
```

```python
pd.merge(df3,df4,how='left')
```
